```python
train_X = train_preprocessed.drop(columns='Stay')
```

```python
test_X = test_preprocessed.drop(columns='Stay')
```

```python
train_y = train_preprocessed['Stay']
```

```python
train_y = train_y.replace({"0-10": 1, "11-20": 2, "21-30": 3,"31-40": 4,"41-50" : 5,"51-60" : 6,"61-70" : 7, "71-80" : 8 , "81-90" : 9, "91-100": 10,"More than 100 Days" : 11 })



```

```python
train_y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

sc.fit(train_X)
sc.fit(test_X)
```

```python
X_scaled_train = pd.DataFrame(data = sc.transform(train_X), columns = train_X.columns)

X_scaled_train.head()
```

```python
X_scaled_test = pd.DataFrame(data = sc.transform(test_X), columns = test_X.columns)

X_scaled_test.head()
```

```python
from sklearn.metrics import classification_report,confusion_matrix,plot_roc_curve,accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
def metrics(y_true,y_pred):
    print('Confusion Matrix:\n', confusion_matrix(y_true, y_pred))
    print('\n\nAccuracy Score:\n', accuracy_score(y_true, y_pred))
    print('\n\nClassification Report: \n', classification_report(y_true, y_pred))
```

```python

lg = LogisticRegression(max_iter = 1000)

lg.fit(X_scaled_train, train_y)
```

```python
train_pred = lg.predict(X_scaled_train)

test_pred = lg.predict(X_scaled_test)
```

```python
print('Train Metrics')
metrics(train_y, train_pred)
```
