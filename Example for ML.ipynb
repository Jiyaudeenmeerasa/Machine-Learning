```python
import matplotlib.pyplot as plt
```

```python
x=[[6],[8],[10],[14],[18],[21]]
y=[[7],[9],[13],[17.5],[18],[24]]
```

```python
plt.figure()
```

```python
plt.title('pizza size prediction')
```

```python
plt.xlabel('Diameter in inches')
plt.ylabel('prize in dollars')
plt.plot(x,y,'*')
plt.axis([0,25,0,25])
plt.grid(True)
plt.show()
```
